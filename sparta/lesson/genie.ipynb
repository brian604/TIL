```python
import requests 
import bs4
import pymongo
client = pymongo.MongoClient("localhost", 27017)
music = client.sparta.music
```

```python
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}
# Need to use header so web scraping does not yield bot alert (security)
url = 'https://www.genie.co.kr/chart/top200?ditc=D&rtm=N&ymd=20190908'
response = requests.get(url, headers = headers)
soup = bs4.BeautifulSoup(response.text,'html.parser')
#print(soup.select('table.list-wrap td'))
# he = soup.select('.number') # ranking, but will it be necessary?
```

## Selector
#body-content > div.newest-list > div > table > tbody > tr:nth-child(1) > td.info > a.artist.ellipsis
#body-content > div.newest-list > div > table > tbody > tr:nth-child(1) > td.info > a.title.ellipsis
#body-content > div.newest-list > div > table > tbody > tr:nth-child(1) > td.check > input

```python
ranking=soup.select('table.list-wrap td')
num = 1
#song == song.select_one('td.check') --> False 
genie190908= []
for song in ranking:
    title= song.select_one('td.info > a.title.ellipsis')
    artist= song.select_one('td.info > a.artist.ellipsis')
    ranking_num = num
    if title is not None:
        real_title = title.text
        genie_one = {'Ranking': ranking_num,
                 'Title': real_title.lstrip(),
                 'Artist': artist.text}
        #music.insert_one(genie_one) <- Into the DB
        num+=1
        for i in range(1,ranking_num):
            genie190908.append(genie_one)
        print(genie_one)
print(genie190908)
```

## Saving all genie top 50 from 20190908 to 20200206

```python
import requests 
import bs4
import datetime
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}
url = 'https://www.genie.co.kr/chart/top200?ditc=D&rtm=N&ymd='
ymd = range(20190909,20200207) # Date로 변환 필요 
date_url = []
for i in ymd:
     date_url.append(url + str(i))
```

```python
for i in date_url[1:3]:
    response = requests.get(i, headers = headers)
    soup = bs4.BeautifulSoup(response.text,'html.parser')
    ranking=soup.select('table.list-wrap td')
    print(ranking)
    
```
